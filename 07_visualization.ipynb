### boxplot的盒体的中间线是medium中位数，盒体上下线是medium的+-25%对应的值(就是25%位和75%位)，盒体往上下延伸出去的是胡须，盒体上下线到中位数线的距离是IQR，在胡须终点处有一个小横杠，代表这里距离盒体的上下线已经有了1.5个IQR，之外的值会特别用点表明出来，代表异常值，看异常值可以推断出数据的离散程度。

### showmeans = True显示平均数，meanline = '.'设定平均数的格式，只有先写出showmeans = True才能设定平均数线的格式，之间的实线是medium中位数不是平均数

### 从上面的box可以看出两组数据的medium都是位于0，盒子的上下限距离中位数的距离基本上相等，mean线也在平均数0的位置，outer异常值不多，代表离散程度低，满足平均数为0的标准正态分布的表现形式。

### plt.setp(ax, xticklabels=['1st', '2nd'])通过[]的方式写出每个箱图对应的label。

```python
def func(x):
    return 0.5 * np.exp(x) + 1
a, b = 0.5, 1.5
x = np.linspace(0, 2)
y = func(x)
Ix = np.linspace(a, b)
Iy = func(Ix) # <6>
verts = [(a, 0)] + list(zip(Ix, Iy)) + [(b, 0)]
```

```python
from matplotlib.patches import Polygon
fig, ax = plt.subplots(figsize=(10, 6))
plt.plot(x, y, 'b', linewidth=2)
plt.ylim(ymin=0)
poly = Polygon(verts, facecolor='0.7', edgecolor='0.5')
ax.add_patch(poly)
plt.text(0.5 * (a + b), 1, r'$\int_a^b f(x)\mathrm{d}x$',
         horizontalalignment='center', fontsize=20)
plt.figtext(0.9, 0.075, '$x$')
plt.figtext(0.075, 0.9, '$f(x)$')
ax.set_xticks((a, b))
ax.set_xticklabels(('$a$', '$b$'))
ax.set_yticks([func(a), func(b)])
ax.set_yticklabels(('$f(a)$', '$f(b)$'))
# plt.savefig('../../images/ch07/mpl_19')
```

plt.text是在图像上面加文字，0.5 * (a + b)是文字所在的大致位置，之后是数学表达式(要用$$夹起来)或者是要写的文字，horizontalalignment='center'水平居中对齐，fontsize=20大小。

plt.figtext(0.9, 0.075, '$x$')设定坐标轴表达式的位置和内容

set_xticks((a, b))和set_xticklabels是设定在坐标轴上的点/刻度的label的表达形式

```python
Ix
```

### 逻辑：先定义一个数学表达式并且命名为func(x)，赋予x以（0，2）linespace取出的50个点，y是这50个0到2的点对应的y值组成的数组。Ix用linespace取出50个（0.5，1.5）的点，将数组Ix过func(x)计算出数值Iy，用zip将Ix和Iy一一对应的组织起来，并且list化，然后添上（0.5, 0）和（1.5,0）一头一尾的x在X轴上的两个点对应的xy坐标。(这样才能构成完整的四边形，ploygon画图形需要每个边的顶点的坐标)
### 因此，0到2用来画函数线段，而0.5到1.5是函数线段的一段，用来做定积分的图像（阴影部分面积）。

linspace(a, b)在a和b之间取数，默认取50个点。

```python
fig, ax = plt.subplots(figsize=(10, 6))
plt.plot(x, y, 'b', linewidth=2)
plt.ylim(ymin=0)
```

单独的线图

```python
from matplotlib.patches import Polygon
fig,ax = plt.subplots(figsize = (10, 6))
poly = Polygon([(0.5, 0), (0.6, 0), (0.6, func(0.6) - 1), (0.5, func(0.5) - 1)], facecolor='0.7', edgecolor='0.5')
ax.add_patch(poly)
```

#### 画出阴影部分，需要底部的两个点和上面的两个点，基于微积分的原理，上面点无限细分后相加(相连)就是定积分所要面积的图像。

func(0.6) - 1是因为polygon默认大小在0到1之间，为了能够在画幅里面看到画出图像的情况就直接-1。

facecolor是里面填充的颜色，edgecolor是边缘的颜色。

```python
fig,ax = plt.subplots(figsize = (10, 6))
poly = Polygon([(0.5, 0), (0.6, func(0.6) - 1), (0.6, 0) , (0.5, func(0.5) - 1)])
ax.add_patch(poly)
```

点的顺序很重要，polygon的原理就是从第一个点出发一直连到最后一个点，然后对之间的平面填充颜色

## Static 3D Plotting立体3D图

```python
strike = np.linspace(50, 150, 24)
```

```python
ttm = np.linspace(0.5, 2.5, 24)
```

```python
strike, ttm = np.meshgrid(strike, ttm)
```

meshgrid相当于是排列组合出所有strike和ttm构成的点组成的矩阵，使两者存在对应关系，位置相同。

例：x轴(0, 0.5),y轴(0.5, 1.5, 2.5),组成矩阵会有2* 3 = 6种组合， meshgrid就是产生出这6种组合，然后再把x值和y值拆开，各自组成新的list；

新的list会有重复，但是于对应的另一个list存在位置上的对应，所以在函数要用到x和y的时候可以直接引用，从而取尽矩阵里面的所有点

```python
ttm[:2].round(1)
```

```python
strike[:2].round(1)
```

```python
ttm.shape
```

```python
iv = (strike - 100) ** 2 / (100 * strike) / ttm
```

```python
iv[:5, :3]
```

```python
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure(figsize=(10, 6))
ax = fig.gca(projection='3d')
surf = ax.plot_surface(strike, ttm, iv, rstride=2, cstride=2,
                       cmap=plt.cm.coolwarm, linewidth=0.5,
                       antialiased=False)
ax.set_xlabel('strike')
ax.set_ylabel('time-to-maturity')
ax.set_zlabel('implied volatility')
fig.colorbar(surf, shrink=0.5, aspect=5);
# plt.savefig('../../images/ch07/mpl_20')
```

ax = fig.gca(projection='3d')建立一个包含3个轴的3D立体画幅

ax.plot_surface是画3D的面图，scatter是点图

rstride和cstride是画出的曲面上的grid的数量。越大越稀疏。

antialiased是否需要平滑化，False会看出来一定的锯齿。

aspect=5是colorbar的参数，就是colorbar的宽度，shrink=0.5是长度。

```python
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection='3d')
ax.view_init(30, 60)
ax.scatter(strike, ttm, iv, zdir='z', s=25,
           c='b', marker='^')
ax.set_xlabel('strike')
ax.set_ylabel('time-to-maturity')
ax.set_zlabel('implied volatility');
# plt.savefig('../../images/ch07/mpl_21')
```

第二种建立3D画幅的方法：ax = fig.add_subplot(111, projection='3d')，通过subplots

view_init是设置3D图的观看角度，（30， 60）是向下旋转30度，向左选择60度，这是最适合观看到全部点的角度。

zdir='z'是设定作为基底的平面，z是以x_y平面为基底，往z的平面画.______ c是color颜色 ——————marker是记号^.

原理就是股价大体上是服从正态分布的，所以绝大多数的正常数值都集中在medium中位数的两侧，如果出现超出布林通道的情况(outer异常值)，有重大的股市调整会出现，具体是利好还是下跌没有科学的规定来绝对判定出来。
